<h2>Moons</h2>

```python
from sklearn.datasets import make_moons
from matplotlib import pyplot
from pandas import DataFrame
%matplotlib inline

# Generate 2d classification dataset
X, y = make_moons(n_samples=500, noise=0.1)

# Scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
from math import ceil
import numpy as np

x0_mean = np.mean(X[:,0])
x1_mean = np.mean(X[:,1])
x0_min = np.min(X[:,0])
x1_min = np.min(X[:,1])
x0_range = (np.max(X[:,0]) - np.min(X[:,0]))
x1_range = (np.max(X[:,1]) - np.min(X[:,1]))
shift = np.max([x0_min, x0_min])
r = np.max([x0_range, x1_range])

print("x0_mean =", x0_mean)
print("x1_mean =", x1_mean)
print("x0_min =", x0_min)
print("x1_min =", x1_min)
print("shift =", shift)
print("range =", r)

rows = int(ceil(X.shape[0] / 8))

for row in range(0,rows):
    for col in range(0,8):
        i = row * 8 + col
        if i < X.shape[0]:
            x0 = (X[i][0] - shift) / r
            x1 = (X[i][1] - shift) / r + 0.08
            print("[{0:.3f},{1:.3f}], ".format(x0, x1), end="")
    print("")
```

<h2>Blobs</h2>

```python
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot
from pandas import DataFrame
%matplotlib inline

# Generate 2d classification dataset
X, y = make_blobs(n_samples=800, centers=5, n_features=2, cluster_std=1.0, random_state=51)
# Scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green', 3:'yellow', 4:'grey'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
from math import ceil
import numpy as np

x0_mean = np.mean(X[:,0])
x1_mean = np.mean(X[:,1])
x0_min = np.min(X[:,0])
x1_min = np.min(X[:,1])
x0_range = (np.max(X[:,0]) - np.min(X[:,0]))
x1_range = (np.max(X[:,1]) - np.min(X[:,1]))
shift = np.max([x0_min, x0_min])
r = np.max([x0_range, x1_range])

print("x0_mean =", x0_mean)
print("x1_mean =", x1_mean)
print("x0_min =", x0_min)
print("x1_min =", x1_min)
print("shift =", shift)
print("range =", r)

rows = int(ceil(X.shape[0] / 8))

for row in range(0,rows):
    for col in range(0,8):
        i = row * 8 + col
        if i < X.shape[0]:
            x0 = (X[i][0] - shift) / r + 0.15
            x1 = (X[i][1] - shift) / r + 0.1
            print("[{0:.3f},{1:.3f}], ".format(x1, x0), end="")
    print("")
```
